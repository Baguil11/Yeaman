Interpolation is when you find the values between 2 or more points given a function. You can also find the function given the points.


Linear Interpolation:  y(x)=yi+[(yi+1−yi)(x−xi)]/(xi+1−xi).  This assumes that the point you are trying to find is along the line and in between two points. Here the interp1d function is used to interpolate the values given from the x and y array. The 1 references how this is a power of one function and not exponential or cubic. The y_hat gives the interpolated value.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)
y_hat = f(1.5)
print(y_hat)
```

Cubic Interpolation: The interpolation assumes the standard cubic function --> Si(x)=aix3+bix2+cix+di. The first task is to find the coefficients a,b,c and d for the cubic equation. Cubic fnctions require a minimum of 4 equations, one for each coefficient. The coefficients can then be solved through a matrix.

We know cubic functions intersect data on left and right points:
Si(xi) == yi
Si(xi+1)==yi+1

The code below describes a cubic spline output.



```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = [1, 2, 3, 4, 5]
y = [3, 10, 5, 1, 6]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 5, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.show()
```
