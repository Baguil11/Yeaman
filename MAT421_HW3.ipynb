Lagrange polynomials differ from cubic splines in that the Lagrange polynomials finds a polynomial that goes through each point rather than just a cubic polynomial in between two sets of points. Essentially, the property: L(x) = y shold be true for every 'x'. The Lagrange Basis plynomials accounts for each point and the sum of these polynomials give the Lagange polynomial. Below is code that demonstrates how the Basis polynomials can be sued to find the Lagrange polynomial.

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = [0, 1, 2, 3, 4]
y = [1, 3, 2, 2, 3]
P1_coeff = [1,-1.5,.5, 1, .5]
P2_coeff = [0, 2,-1, 3, -2]
P3_coeff = [0,-.5,.5, 1, .5]
P4_coeff = [0,.5,.5, 1, -1]
P5_coeff = [0,1,-.5, 2, 1]

P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)
P4 = poly.Polynomial(P4_coeff)
P5 = poly.Polynomial(P5_coeff)

x_new = np.arange(-1.0, 5, .1)

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')
plt.plot(x_new, P4(x_new), 'g', label = 'P4')
plt.plot(x_new, P5(x_new), 'g', label = 'P5')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

L = P1 + 3*P2 + 2*P3 + 2*P4 + 3*P5

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
